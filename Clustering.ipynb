```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use("ggplot")
plt.rcParams["figure.figsize"]=(12,8)
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=150,n_features=2, centers=4, random_state=1)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4,init="k-means++", n_init=10, random_state=1)
kmeans.fit(X)
```

```python
labels=kmeans.labels_ #после обучения кминс, можем получить метки кластеров
```

```python
plt.scatter(X[:,0],X[:,1], c=labels)
```

##### Теперь узнаем какое К самое наилучшее для нашего датасета, для этого мы используем метод локтя 

```python
crit=[]
for k in range(2,8):
    kmeans=KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    
    crit.append(kmeans.inertia_)
```

```python
crit
```

```python
plt.plot(range(2,8), crit)
```
